#My 1st Python Prog

```python
a = [10, 20, 30, 40, 50, 60, 70, 80]
b = [9, 10, 11, 12, 13, 14, 15, 16]
c = [x + y for x, y in zip(a, b)]
d = [x * y for x, y in zip(a, b)]
e = [x - y for x, y in zip(a, b)]
f = [x / y for x, y in zip(a, b)]
g = [x ** y for x, y in zip(a, b)]
print(c)
print(d)
print(e)
print(f)
print(g)
```

#Tables from 1 to 100

```python
i = 1
while i <= 100:
    print("TABLE OF", i)
    j = 1
    while j <= 10:
        print(i * j, end=" ")
        j += 1
    print("\n")
    for j in range(1, 11):
        print(f"{i} x {j} = {i * j}")
    print(f'End of table of {i}\n')
    i += 1
```
